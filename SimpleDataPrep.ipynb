```python
import numpy as np
#in case we need to repeat experiment
np.random.seed(255)

import pandas as pd
pd.options.display.max_rows = 25

import matplotlib.pyplot as plt
plt.style.use('classic')

from IPython.display import display, HTML

import seaborn as sns
sns.set()
```

```python
natmp= np.array([['Col1','Col2','Col3','Col4', 'Col5'],
                 ['1',1,2,'01', 1],
                 ['2',2,3,'10',2],
                 ['3','U','03', 4,3],
                 ['4',1,4,'10', 1],
                 ['U','N',6,'05',2],
                 ['6',3,7, 'U',3],
                 ['N',1,'N','06',2],
                 ['8',2,3,'08',3],
                 ['9','U','U','07',3],
                 ['10',2,'U','05',3],
                 ['11',3,3,'10',3],
                 ['12',1,7,'04',2],
                 ['13',3,8,'01',2],
                 ['14',2,5,'05',1],
                 ['15',1,4,'08',2],
                 ['16',1,3,'10',3],
                 ['17',2,2,'10',1],
                 ['18',1,5,'04',3]])

df = pd.DataFrame(data=natmp[1:,0:], columns=natmp[0,0:])
display(df)
print(df.index.name)
print(df.columns.name)
print(df.columns)
```

```python
for col in df.columns:
    impute = df[col].astype('str').str.contains('[^0-9]')
    df[col][impute] = df[col][impute].apply(lambda x: np.nan)
df_no_missing = df.dropna()
df_no_missing = df_no_missing.reset_index(drop=True)
df_no_missing
```

```python
df_no_missing.to_csv('data01_simple.csv', encoding='utf-8', index=False)
```
